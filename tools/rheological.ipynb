```python
fig,ax = plt.subplots(1,1, figsize=(9,3))

ax.axvline(0.3, color='r', label='0.3', ls='dashed')
ax.scatter(x,y, color='k', label="Their figure 1")

phi_crit = 0.4

phi_arr = np.linspace(0.0, 1.0, 100)
mod_arr = [eval_rheoparam(p, 'shear', phi_crit) for p in phi_arr]
ax.plot(phi_arr, mod_arr, label="%f"%phi_crit, color='b')
ax.axvline(x=phi_crit, color='b', ls='dashed')

ax.legend()
ax.set_yscale("log")
ax.set_xlabel("Melt fraction")
ax.set_ylabel("Shear modulus [Pa]")

plt.show()
```
